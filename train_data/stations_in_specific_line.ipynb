```python
def line_name_cd(line_name):
    matched_lines = line_df[line_df['line_name'] == line_name]['line_cd']
    if len(matched_lines) != 1:
        raise ValueError('There is no lines such as {}'.format(line_name))
    return int(matched_lines)     
```
